```python
model = tf.compat.v1.global_variables_initializer()
data = read_csv('dhpricedata.csv', sep=',')
xy = np.array(data, dtype=np.float32)
print(xy)
```

```python
x_data = xy[:, 1:-1]
y_data = xy[:, [-1]]
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x = tf.placeholder(tf.float32, shape=[None, 4])
y = tf.placeholder(tf.float32, shape=[None, 1])
w = tf.Variable(tf.random_normal([4,1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
# value initialization part
```

```python
hypothesis = tf.matmul(x, w) + b
# Establish hypotheses, matrix operations
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
# Set cost function
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0000005)
train = optimizer.minimize(cost)
# Optimization function settings (learning rate: 0.000005) The learning rate is the degree of data refinement, but it is set differently depending on the type
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
# Creating a session value and performing learning according to the session
```

```python
for step in range(100001):
  cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={x: x_data, y: y_data})
  if step % 500 == 0:
    print("#", step, "loss cost:", cost_)
    print("-cabbage price: ", hypo_[0])
# Process of checking progress every 500 steps
```

```python
saver = tf.train.Saver()
save_path = saver.save(sess, "./saved.cpkt")
print("Saved the trained model")
# The process of saving a model
```
